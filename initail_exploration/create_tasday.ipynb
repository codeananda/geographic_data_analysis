```python
import pandas as pd
from pathlib import Path

def get_lon_lat(string):
    """
    Given a string of the form 'lon=X lat=Y' for some floats X and Y each to 2 decimal places, 
    return the tuple (X, Y)
    """
    space = string.find(' ')
    lon = float(string[4:space])
    lat = float(string[space+5:])
    return lon, lat


def process_csv(csv_path):
    csv_path = Path(csv_path)
    df = pd.read_csv(csv_path, parse_dates=[0], names=['date', 'temperature'], skiprows=1)
    
    lon, lat = get_lon_lat(csv_path.parent.parts[-1])
    df['lon'] = pd.Series([lon for i in range(len(df))])
    df['lat'] = pd.Series([lat for i in range(len(df))])
    
    df = df.set_index(['lon', 'lat', 'date'])
    return df
```

```python
csv_paths = list(Path('Cells_tasday').glob('*/*.csv'))

# This is the same as the func definition below
lambda x: int(csv_paths[x].stem[:8])
# This is the same as the lambda above
def my_func(x):
    return int(csv_paths[x].stem[:8])

# Not sure why I sorted them, would have made more sense to not sort them so that each cell was done
# at once. After sorting, it meant that all the times were done together i.e. all the 2006-2010 csvs
# were processed first for every cell, then all the 2020 etc.
csvs_sorted = sorted(csv_paths, key=lambda x: int(x.stem[:8]))
```

```python
# Also use this function to create tasmax and tasmin 
# (I think we stopped using tasday later on in the project though)
def create_tasday(list_of_paths):
    for index, csv_path in enumerate(list_of_paths):
        df = process_csv(csv_path)
        if index == 0:
            df.to_csv('tasday.csv')
            print(f'List index {index}, successfully added {csv_path} to tasday.csv')
        else:
            df.to_csv('tasday.csv', mode='a', header=False)
            print(f'List index {index}, successfully added {csv_path} to tasday.csv')
```
